```
def dataset_build_util(src_dir_path):
  for root, dirs, file_names in os.walk(src_dir_path):
    images_found = False  
    if 'control' in root:
      dst_dir_path = '/content/Plant_leave_diseases_dataset_without_augmentation/Wheat___healthy' + '/'
      images_found = True
    elif 'deficiency' in root:
      dst_dir_path = '/content/Plant_leave_diseases_dataset_without_augmentation/Wheat___Nitrogen_deficiency' + '/'
      images_found = True
    elif 'diseased' in root:
      dst_dir_path = '/content/Plant_leave_diseases_dataset_without_augmentation/Wheat___Leaf_rust' + '/'
      images_found = True
    if images_found:
      for file_name in file_names:
        img = Image.open(root + '/' + file_name)
        path = dst_dir_path + file_name
        img.save(path)
```

```
dataset_build_util('/content/Ndeficient')
dataset_build_util('/content/WheatLeafRust')
```

### Data preprocessing

```
BATCH_SIZE = 128
```

```
IMAGE_SIZE = 224
```

```
datagen = ImageDataGenerator(rotation_range=90,
                             width_shift_range=0.2,
                             height_shift_range=0.2,
                             brightness_range=[0.5, 1.5],
                             shear_range=0.1,
                             zoom_range=0.2,
                             horizontal_flip=True,
                             vertical_flip=True,
                             rescale=1./255,
                             validation_split=0.2)
```

```
train_generator = datagen.flow_from_directory(directory='/content/Plant_leave_diseases_dataset_without_augmentation', 
                                              target_size=(IMAGE_SIZE, IMAGE_SIZE), 
                                              color_mode='rgb', 
                                              class_mode='categorical', 
                                              batch_size=BATCH_SIZE, 
                                              shuffle=True, 
                                              subset='training')

test_generator = datagen.flow_from_directory(directory='/content/Plant_leave_diseases_dataset_without_augmentation', 
                                              target_size=(IMAGE_SIZE, IMAGE_SIZE), 
                                              color_mode='rgb', 
                                              class_mode='categorical', 
                                              batch_size=BATCH_SIZE, 
                                              shuffle=False, 
                                              subset='validation')
```

### Build the model

```
vgg = VGG16(input_shape=[IMAGE_SIZE, IMAGE_SIZE, 3], weights='imagenet', include_top=False)
```

```
vgg.summary()
```

```
model = Sequential()
```

```
for layer in vgg.layers:
  model.add(layer)
```

```
for layer in model.layers:
  layer.trainable = False
```

```
DATA_DIR = '/content/Plant_leave_diseases_dataset_without_augmentation'
CATEGORIES = os.listdir(path=DATA_DIR)
```

```
model.add(Flatten(name='flatten'))
model.add(Dense(units=len(CATEGORIES), activation='softmax'))
```

```
model.summary()
```

```
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

### Train the model

##### First 5 epochs

```
EPOCHS = 10
```

```
history = model.fit(x=train_generator,
                    epochs=EPOCHS,
                    steps_per_epoch=train_generator.samples // train_generator.batch_size)
```

```
print('Epoch-wise training loss and training accuracy:\n')
print(history.history)
```

```
results = model.evaluate(x=test_generator)
```

```
print(f'Validation Loss: {results[0]}')
print(f'Validation Accuracy: {results[1]}')
```

```
model.save(file_path)
```
